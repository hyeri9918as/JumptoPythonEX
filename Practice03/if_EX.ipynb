# if 문

#### if 문은 주어진 조건을 판단한 후 그 상황에 맞게 상황을 처리할 때 사용 

---


## if 문 사용 예 

### 돈이 있으면 택시를 타고, 없으면 걸어가라

```python
money = True            # 돈이 있다라는 의미 

if money :              # 돈이 있을때 
    print("택시를 타고 가라")
else :                  # 돈이 없을 때 
    print("걸어가라")
```

## 조건문의 정의

#### if 조건문에서 조건문이란, 참과 거짓을 판단하는 문장을 말한다. 

```python
money = True     # 조건문 
if money : 
    print("택시를 타고 가라")
```

#### ● 비교연산자 

조건문에 사용되는 비교연산자 

```python
from IPython.display import Image
Image("IMG/comparison_Operator.png")
```

#### ● 비교연산자 사용 예 

#### > 연산자 

```python
x = 3
y = 2
x > y      # 3 > 2 / 참
```

#### < 연산자 

```python
x = 3
y = 2
x < y      # 3 < 2 / 거짓 
```

#### == 연산자 

```python
x = 3
y = 2
x == y      # 3 == 2 / 값이 같지 않으므로 거짓 
```

#### != 연산자 

```python
x = 3
y = 2
x != y      # 3 != 2 / 값이 같지 않으므로 참 
```

## 연습문제 

### 만약 3000원 이상의 돈을 가지고 있으면 택시를 타고 그렇지 않으면 걸어가라.

```python
money = 2000       # 조건문 : 돈을 2000원 가지고 있음 

if money >= 3000 :
    print("택시를 타고 가라")     # 돈이 3000원 이상일 때 
else : 
    print("걸어 가라")            # 돈이 3000원 미만일 때 
```
