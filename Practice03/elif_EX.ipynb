# elif 문 

#### 조건문이 두 개 이상의 조건을 가지고 있을 때 사용하면 효율적으로 코드를 운용할 수 있다. 

---

## 주머니에 돈이 있으면 택시를 타고, 주머니에 돈은 없지만 카드가 있으면 택시를 타고, 돈도 없고 카드도 없으면 걸어가라. 

#### if / else 문 코드

```python
pocket = ['paper', 'cellphone']
card = True

if 'money' in pocket :          # 주머니 리스트에 'money'가 있다면 
    print("택시를 타고 가라")
else :                          # 리스트에 'money'가 없지만 
    if card :                   # 카드가 있는 경우 
        print("택시를 타고 가라")   
    else :                      # 카드도 없는 경우 
        print("걸어가라")
```

#### elif 문 코드 

```python
pocket = ['paper', 'cellphone']
card = True

if 'money' in pocket :          # 주머니 리스트에 'money'가 있다면 
    print("택시를 타고 가라")
elif card :                     # 주머니 리스트에 'money'가 없고 / 카드가 있다면 
    print("택시를 타고 가라")
else :                          # 둘다 없으면 
    print("걸어가라")
```
