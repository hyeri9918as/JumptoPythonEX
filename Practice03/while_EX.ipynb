## while 문 사용 예 

#### 나무를 열번 찍어서 넘어가게 하는 상황 설정 

```python
treeHit = 0           # 변수값 디폴트 0 

while treeHit < 10:                   # 나무 찍는 횟수가 10보다 작은 동안 반복 
    treeHit = treeHit + 1             # 나무를 찍은 횟수 1씩 증가 
    print("나무를 %d번 찍었습니다." % treeHit) 
    
    if treeHit == 10:                 # 나무를 10번 찍으면 
        print("나무 넘어갑니다.")
```

#### while 문의 조건문은 treeHit < 10이다. 즉, treeHit이 10이 되면 조건문이 거짓이 되므로 if문으로 넘어가지게 되고, while 문을 빠져나갈 수 있게 된다.

## while 문 선택지 입력 예제  

### ① 변수에 선택지 문자열을 입력 

```python
prompt = """
    1. Add
    2. Del 
    3. List 
    4. Quit 

Enter number: """
```

### ② number 변수에 0을 대입 (디폴트 값) 

#### number가 4가 아닌 동안 promot를 출력하고 사용자로부터 번호를 입력받는다. 

```python
number = 0            # 번호를 입력받을 변수 (기본값) 
while number != 4 :   # 입력받은 번호가 4가 아닌 동안 반복한다. 
    print(prompt)
    number = int(input())   # 번호를 입력받는 창 
```
